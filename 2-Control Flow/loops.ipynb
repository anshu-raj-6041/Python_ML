```python
# while loop
count = 0

while (count <= 5):
    print(count)
    count += 1
```

```python
count = 0

while (count % 2 == 0):
    print(count)
    # count = count + 1
```

```python
# Loop Control Statement
# break
for i in range(10):
    if (i==5):
        break
    print(i)
```

```python
# continue
# skip the current iteration and continues with the next
# odd print krna hai
for i in range(10):
    if (i%2 == 0):
        continue
    print(i)
```

```python
# pass
# The pass statement is a null operation, it does nothing
for i in range(5):
    if (i==3):
        pass
    print(i)
```

```python
# Nested loop
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
# calculate sum of first n natural no using while loop
n = 10
sum = 0
count = 1

while (count <= n):
    sum += count
    count += 1

print("Sum => ", sum)
```

```python
# using for loop
n = 10
sum = 0

for i in range(11):
    sum += i
print(sum)
```

```python
# Ex- Display prime no b/w 1 and 100
for num in range(1, 101):
    if (num > 1):
        for i in range(2, num):
            if (num%i==0):
                break
        else:
            print(num)
```
